```python
#BUBBLE SORT

import matplotlib.pyplot as plt
from IPython.display import clear_output
import numpy as np

amount = 15

lst = np.random.randint(0, 100, amount)
x = np.arange(0, amount, 1)
n = len(lst)

# for color list to be passed as per the current index
def plot_bars(lst, highlight_index):
    colors = ['blue' if i == highlight_index else 'red' for i in range(len(lst))]
    plt.bar(x, lst, color=colors)

plt.bar(x, lst, color='red')
plt.draw()
# plt.pause(1)  # Add a pause before starting the sorting process

for i in range(n):
    for j in range(0, n-i-1):

        if lst[j] > lst[j+1]:
            lst[j], lst[j+1] = lst[j+1], lst[j]

        # Update the heights of the bars with a highlighted moving bar
        plot_bars(lst, highlight_index=j)
        plt.draw()
        plt.pause(0.001)  # Add a small pause to see the animation
        clear_output(wait=True)  # Clear the output after each iteration

# Display the final sorted plot
plot_bars(lst, highlight_index=-1)  # Highlight no bar (i.e., all bars are the same color)
plt.show()
```

```python
# MERGE SORT

import matplotlib.pyplot as plt
import numpy as np
import time

def merge_sort_visualized(arr, left, right):
    if left < right:
        mid = (left + right) // 2
        
        merge_sort_visualized(arr, left, mid)
        merge_sort_visualized(arr, mid+1, right)
        
        merge(arr, left, mid, right)

        # Visualizing the sorting process on a single bar graph
        clear_output(wait=True)
        plt.bar(np.arange(len(arr)), arr, color=['blue' if i == right else 'red' for i in range(len(arr))])
        plt.draw()
        plt.pause(0.1)

def merge(arr, left, mid, right):
    left_arr = arr[left:mid + 1]
    right_arr = arr[mid + 1:right + 1]
    i, j, k = 0, 0, left

    while i < len(left_arr) and j < len(right_arr):
        if left_arr[i] <= right_arr[j]:
            arr[k] = left_arr[i]
            i += 1
        else:
            arr[k] = right_arr[j]
            j += 1
        k += 1

    while i < len(left_arr):
        arr[k] = left_arr[i]
        i += 1
        k += 1

    while j < len(right_arr):
        arr[k] = right_arr[j]
        j += 1
        k += 1

def main():
    arr = [38, 27, 43, 3, 9, 82, 10]
    plt.figure(figsize=(8, 4))
    plt.bar(np.arange(len(arr)), arr, color='blue')
    plt.draw()
    plt.pause(1)

    merge_sort_visualized(arr, 0, len(arr) - 1)
    
    # display the final sorted array in a single bar graph
    clear_output(wait=True)
    plt.bar(np.arange(len(arr)), arr, color='red')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
# QUICK SORT

import matplotlib.pyplot as plt
import random

# Quick Sort algorithm
def quick_sort(arr, low, high):
    if low < high:
        pivot_index = partition(arr, low, high)
        quick_sort(arr, low, pivot_index - 1)
        quick_sort(arr, pivot_index + 1, high)

def partition(arr, low, high):
    pivot = arr[high]
    pivot_index = low
    for i in range(low, high):
        if arr[i] < pivot:
            arr[i], arr[pivot_index] = arr[pivot_index], arr[i]
            pivot_index += 1
            draw(arr)
    arr[pivot_index], arr[high] = arr[high], arr[pivot_index]
    draw(arr)
    return pivot_index

# Visualization function
def draw(arr):
    clear_output(wait = True)
    plt.clf()  # Clear the previous plot
    plt.bar(range(len(arr)), arr, color='blue')
    plt.pause(0.05)  # Pause to allow visualization to update

# Main function
def main():
    num_elements = 20
    arr = [random.randint(10, 100) for _ in range(num_elements)]

    plt.figure(figsize=(8, 6))
    plt.ion()  # Turn on interactive mode for dynamic updates
    plt.title("Quick Sort Visualization")

    draw(arr)
    quick_sort(arr, 0, len(arr) - 1)

    plt.ioff()  # Turn off interactive mode after sorting
    plt.show()

if __name__ == "__main__":
    main()
```
